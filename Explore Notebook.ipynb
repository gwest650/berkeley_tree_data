### Creating ID Column

```python
# create id column
berkeley_trees = berkeley_trees.reset_index()
```

The reason why we want to create an index column is that it helps identify each row and if we need to call on a certain row, we can use the index number.

```python
berkeley_trees.head()
```

```python
berkeley_trees.rename(columns={'index': 'id'}, inplace=True)
```

```python
berkeley_trees.head()
```

## Get Total Trees By Zip Code


```python
berkeley_trees.groupby(['ZIP']).count()
```

```python
tree_count_by_zip = berkeley_trees.groupby(['ZIP']).count()
```

```python
tree_count_by_zip
```

```python
tree_count_by_zip = tree_count_by_zip[['id']].reset_index()
```

```python
tree_count_by_zip.head()
```

```python
tree_count_by_zip.columns = ['zip', 'count']
tree_count_by_zip
```

```python
alt.Chart(tree_count_by_zip).mark_bar().encode(
    x='zip',
    y='count'
)
```

Based on the data, the zip code with the most trees is 94703, which includes North, South and Central Berkeley and the zip code with the least amount of trees is 94720 which includes parts of the UC Berkeley campus and Southside, however, I am sure that UC Berkeley has its own tree dataset. The second lowest zip code is 94706, which includes Albany, which is quite surprising. 

## Get Tree Density

How evenly spread out are the trees in Berkeley or are they denser in certain communities?

```python
zip_codes = pd.read_csv('berkeley_zip_codes.csv', dtype={'ZIP': object})
```

```python
zip_codes
```

```python
tree_stats_by_zip = pd.merge(
    tree_count_by_zip,
    zip_codes,
    left_on='zip',
    right_on='ZIP',
    how='outer',
    validate='1:1',
)
```

```python
tree_stats_by_zip
```

```python
tree_stats_by_zip['count_by_sqmi'] = tree_stats_by_zip['count'] / tree_stats_by_zip['SQMI'] 
tree_stats_by_zip
```

```python
tree_stats_by_zip.sort_values(by=['count_by_sqmi'], ascending=False)
```

```python
alt.Chart(tree_stats_by_zip).mark_bar().encode(
    x='zip',
    y='count_by_sqmi'
)
```

Based on my observation, it looks like zip codes 94703 and 94702 have the highest tree density per square mile while zip codes 94720 and 94706 have the least tree density per square mile. In further analysis, it would be interesting to see how the count by square mile changes after 2014.
